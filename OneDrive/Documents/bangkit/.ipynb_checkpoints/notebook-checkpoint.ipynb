## **Exploratory Data Analysis (EDA)**

```python
# Menghitung rata-rata PM2.5 untuk setiap kota per tahun
pm25_city_year = combined_df.groupby(['station', 'year']).agg({'PM2.5': 'mean'}).reset_index()
pm25_city_year
```

```python
# melihat korelasi antara tingkat polusi dengan keadaan cuaca
corr_matrix = combined_df[['PM2.5', 'PM10','SO2', 'NO2', 'CO', 'O3','TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']].corr()
corr_matrix
```

```python
# Mencari kota mana yang mengalami tingkat polusi tertinggi selama tahun 2017
year_2017_data = combined_df[combined_df['year'] == 2017]

# Menghitung rata-rata PM2.5 per stasiun untuk tahun 2017
average_pollution_2017 = year_2017_data.groupby('station')['PM2.5'].mean().reset_index()
average_pollution_2017.max()

```

## **Visualization & Explanatory Analysis**

### *Pertanyaan 1*

```python
# Membuat FacetGrid untuk plot per kota
g = sns.FacetGrid(pm25_city_year, col='station', col_wrap=3, height=4, aspect=1.5)

# Menambahkan line plot di setiap facet (kota)
g.map(sns.lineplot, 'year', 'PM2.5', marker="o")

g.set_titles("{col_name}")
g.set_axis_labels("Tahun", "Rata-rata Konsentrasi PM2.5")
g.fig.suptitle('Pola Perubahan Konsentrasi PM2.5 di Setiap Kota per Tahun', y=1.03)

plt.tight_layout()
plt.show()
```

### *Pertanyaan 2*

```python
# Membuat heatmap untuk korelasi
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Korelasi antara PM2.5 dan Variabel Cuaca', fontsize=14)

plt.show()
```

### *Pertanyaan 3*

```python

# Mengatur ukuran plot
plt.figure(figsize=(8, 6))

sns.barplot(data=average_pollution_2017, x='station', y='PM2.5')

plt.title('Rata-rata PM2.5 per Stasiun - Tahun 2017', fontsize=16)
plt.xlabel('Stasiun', fontsize=12)
plt.ylabel('Rata-rata PM2.5', fontsize=12)
plt.xticks(rotation=45)

# Menambahkan keterangan angka di atas setiap bar
for index, row in average_pollution_2017.iterrows():
    plt.text(index, row['PM2.5'] + 1, f'{row["PM2.5"]:.2f}', color='black', 
             ha='center', fontsize=10)
    
plt.tight_layout()
plt.show()
```

#### 2. Apakah ada korelasi antara konsentrasi PM2.5 dan kondisi cuaca ?

- PM10 (0.88):Ada korelasi positif yang sangat kuat (0.88) antara PM2.5 dan PM10. menunjukkan bahwa saat konsentrasi PM10 meningkat, PM2.5 juga cenderung meningkat.

- SO2 (0.48): Korelasi positif moderat (0.48) antara PM2.5 dan SO2. Ini menunjukkan bahwa peningkatan konsentrasi SO2 dapat berhubungan dengan peningkatan PM2.5, meskipun tidak sekuat hubungan dengan PM10.

- NO2 (0.66):Terdapat korelasi positif yang cukup kuat (0.66) antara PM2.5 dan NO2. Ini menunjukkan bahwa konsentrasi PM2.5 cenderung meningkat bersamaan dengan peningkatan NO2.

- CO (0.78):Korelasi positif yang kuat (0.78) antara PM2.5 dan CO. Ini menunjukkan bahwa saat konsentrasi CO meningkat, PM2.5 juga cenderung meningkat

- O3 (-0.15):Korelasi negatif yang lemah (-0.15) antara PM2.5 dan O3. Ini menunjukkan bahwa perubahan konsentrasi O3 tidak memiliki dampak signifikan pada PM2.5.

- TEMP (-0.13):Korelasi negatif yang sangat lemah (-0.13) antara PM2.5 dan TEMP. Ini menunjukkan bahwa perubahan suhu tidak berhubungan secara signifikan dengan konsentrasi PM2.5.

- PRES (-0.02):Korelasi yang sangat lemah dan hampir tidak ada (-0.02) antara PM2.5 dan PRES. Ini menunjukkan bahwa tekanan atmosfer tidak berpengaruh pada PM2.5.

- DEWP (0.22): Korelasi positif yang lemah (0.22) antara PM2.5 dan DEWP. Meskipun ada hubungan, itu tidak cukup kuat untuk diandalkan.

- RAIN (0.18):Korelasi positif yang lemah (0.18) antara PM2.5 dan RAIN. Ini menunjukkan bahwa curah hujan mungkin sedikit terkait dengan konsentrasi PM2.5, tetapi hubungan ini juga tidak kuat.

- WSPM (-0.27): Korelasi negatif yang moderat (-0.27) antara PM2.5 dan WSPM (kecepatan angin). Ini menunjukkan bahwa peningkatan kecepatan angin cenderung berhubungan dengan penurunan konsentrasi PM2.5.ak kuat.
